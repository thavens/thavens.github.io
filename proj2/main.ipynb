```python
def conv2d(im: np.ndarray, kernel: np.ndarray, padding: int = 0):
    # if padded is shape 3 and kernel is 3 then we do index 0 only
    # if padded_im is shape 3 and kernel is 2 then we have index 0, 1 only
    
    # flip the kernel
    kernel = np.copy(kernel[::-1, ::-1])
    
    padded_im = np.pad(im, (padding, padding)) # zero pad
    result_height = padded_im.shape[0] - kernel.shape[0] + 1
    result_width = padded_im.shape[1] - kernel.shape[1] + 1
    result = np.zeros((result_height, result_width))
    for y in range(result_height):
        for x in range(result_width):
            sub_im = padded_im[y : y + kernel.shape[0], x : x + kernel.shape[1]]
            result[y, x] = np.sum(sub_im * kernel)

    return result
```

```python
def verify_conv2d():
    for p in range(3):
        for i in range(5):
            kernel = np.random.uniform(0, 1, (i, i))
            for j in range(i, 10):
                    im = np.random.uniform(0, 1, (j, j))
                    result = conv2d(im, kernel, p)
                    result_baseline = scipy.signal.convolve2d(np.pad(im, (p, p)), kernel, mode="valid")
                    close = np.isclose(result, result_baseline).all()
                    if not close:
                        print(f"p: {p}, i: {i}, j: {j}", )
                        print(kernel)
                        print(im)
                        print(result)
                        print(result_baseline)
                        return kernel, im
    return 1, 1
    
```
